```python
def generate_rare_board(target_event: str) -> chess.Board:
    """
    Generate a board state featuring a specific rare event using Biased Random Walks.
    target_event options: 'en_passant', 'castling', 'promotion', 'pin'
    """
    while True:
        board = chess.Board()
        
        # Limit to 80 moves. If the event hasn't happened by then, 
        # the board is probably locked up. We break and start a new game.
        for _ in range(80):
            if board.is_game_over():
                break

            legal_moves = list(board.legal_moves)

            # ---------------------------------------------------------
            # 1. CHECK FOR SUCCESS: Do we have the event right now?
            # ---------------------------------------------------------
            if target_event == "en_passant" and any(board.is_en_passant(m) for m in legal_moves):
                return board
                
            if target_event == "castling" and any(board.is_castling(m) for m in legal_moves):
                return board
                
            if target_event == "promotion" and any(m.promotion for m in legal_moves):
                return board
                
            if target_event == "pin":
                # Check if any piece belonging to the current player is absolutely pinned to the king
                if any(board.is_pinned(board.turn, sq) for sq in chess.SQUARES if board.color_at(sq) == board.turn):
                    return board

            # ---------------------------------------------------------
            # 2. BIASED WEIGHTS: Nudge the randomizer toward our goal
            # ---------------------------------------------------------
            weights = []
            for m in legal_moves:
                weight = 1.0  # Base probability
                moving_piece = board.piece_at(m.from_square)

                if target_event == "en_passant":
                    # If we see a 2-square pawn push, test if it grants the opponent en passant.
                    if moving_piece and moving_piece.piece_type == chess.PAWN:
                        if abs(m.from_square - m.to_square) == 16:
                            board.push(m)
                            if any(board.is_en_passant(em) for em in board.legal_moves):
                                weight = 1000.0  # FORCE this move to happen!
                            board.pop()

                elif target_event == "promotion":
                    # Give pawn pushes a massive multiplier so they race down the board
                    if moving_piece and moving_piece.piece_type == chess.PAWN:
                        weight = 50.0

                elif target_event == "castling":
                    # Penalize moving the King or Rooks early to preserve castling rights
                    if moving_piece and moving_piece.piece_type in [chess.KING, chess.ROOK]:
                        weight = 0.05

                elif target_event == "pin":
                    # Encourage long-range sliding pieces to open up the board
                    if moving_piece and moving_piece.piece_type in [chess.QUEEN, chess.ROOK, chess.BISHOP]:
                        weight = 5.0
                    # Checks often force the opponent to block, creating instant pins
                    if board.gives_check(m):
                        weight = 20.0

                weights.append(weight)

            # ---------------------------------------------------------
            # 3. MAKE THE MOVE
            # ---------------------------------------------------------
            chosen_move = random.choices(legal_moves, weights=weights, k=1)[0]
            board.push(chosen_move)
```

```python
import json
from tqdm import tqdm

# Import everything from your file
from data_generators.legal_moves_generator import *

# =====================================================================
# 1. UPGRADED KEYWORD MATCHER
# =====================================================================
def sample_by_keyword(board: chess.Board, keywords: list, max_retries: int = 50):
    """
    Takes a board and a LIST of keywords/key sentences. 
    Returns the sample if ANY of the keywords are found.
    """
    for _ in range(max_retries):
        sample = sample_specific_piece(board) 
        if sample is None: 
            return None
        
        answer_text = sample["messages"][2]["content"].lower()
        
        # Check if ANY of the key sentences in our list appear in the text
        if any(kw.lower() in answer_text for kw in keywords):
            return sample
            
    return None

# =====================================================================
# 2. THE PIPELINE CONFIGURATION
# =====================================================================
num_samples = 50000
output_file = "data/rare_legal_moves.jsonl"
rare_events = ["en_passant", "castling", "promotion", "pin"]

# Map events to ALL possible phrases your random.choice() might generate
keyword_map = {
    # Covers all 5 templates in _PINNED_NO_MOVES exactly
    "pin": [
        "absolute pin", 
        "absolutely pinned"
    ],
    
    # Covers all 5 templates in _EN_PASSANT_NOTICES
    "en_passant": [
        "en passant"
    ],
    
    # Standard generation strings
    "castling": [
        "castling is available", 
        "castling options are available"
    ],
    "promotion": [
        "promotion available"
    ]
}

# =====================================================================
# 3. EXECUTION LOOP
# =====================================================================
generated_count = 0

print("Starting pipeline: Generate Rare Board -> Find Key Sentences -> Loop")

with open(output_file, "w", encoding="utf-8") as f:
    
    with tqdm(total=num_samples, desc="Farming Rare Events") as pbar:
        
        while generated_count < num_samples:
            
            # Step 0: Pick the target event & get its list of key sentences
            target_event = rare_events[generated_count % len(rare_events)]
            target_keywords = keyword_map[target_event]
            
            # Step 1: Generate Rare Board
            board = generate_rare_board(target_event)
            
            # Step 2: Generate Good Sample (checking against the LIST of phrases)
            sample = sample_by_keyword(board, target_keywords, max_retries=50)
            
            # Step 3: Loop / Save
            if sample is None:
                continue
                
            f.write(json.dumps(sample, ensure_ascii=False) + "\n")
            
            generated_count += 1
            pbar.update(1)

print(f"✅ Pipeline complete! Saved {generated_count} rare samples to: {output_file}")
```

```python
import torch
import chess
from peft import PeftModel
from typing import Optional

from config import ChessCoachConfig, MaiaConfig
from model import ChessCoach
from maia3.dataset import tokenize_board, get_historical_tokens

# ══════════════════════════════════════════════════════════════════════════════
# 1. Model Loading Function
# ══════════════════════════════════════════════════════════════════════════════

def load_trained_coach(checkpoint_dir: str) -> ChessCoach:
    """
    Loads the base Qwen3 model, injects your trained MaiaProjectors, 
    and applies your trained LoRA weights.
    """
    print("Initializing base ChessCoach...")
    config = ChessCoachConfig()
    
    # Optional: If you want to use Flash Attention 2 during inference for speed
    # config.attn_implementation = "flash_attention_2"
    
    coach = ChessCoach(config)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    # 1. Load the trained MaiaProjectors (The bridge between Maia and Qwen)
    print(f"Loading MaiaProjectors from {checkpoint_dir}...")
    projector_path = f"{checkpoint_dir}/projectors.pt"
    coach.projectors.load_state_dict(
        torch.load(projector_path, map_location=device, weights_only=True)
    )
    
    # 2. Load the Qwen LoRA Adapters via PEFT
    print(f"Loading LoRA adapters from {checkpoint_dir}/lora...")
    coach.llm = PeftModel.from_pretrained(coach.llm, f"{checkpoint_dir}/lora")
    
    # Move everything to GPU and set to evaluation mode
    coach = coach.to(device)
    coach.eval()
    
    print("Model fully loaded and ready for inference!")
    return coach

# ══════════════════════════════════════════════════════════════════════════════
# 2. Inference Function
# ══════════════════════════════════════════════════════════════════════════════

@torch.inference_mode()
def ask_chess_coach(
    coach: torch.nn.Module, 
    board: chess.Board, 
    user_prompt: str, 
    self_elo: int = 1500, 
    oppo_elo: int = 1500,
    max_new_tokens: int = 1024,
    temperature: float = 0.2 # Keep temp low for strict logical evaluation
) -> str:
    """
    Takes a board and a text prompt, processes them through Maia, and returns 
    Qwen's generated CoT and final answer.
    """
    device = next(coach.parameters()).device
    
    # 1. Extract specific moves from the prompt (e.g., "Why is Nf3 bad?")
    safe_ucis = []
    for word in user_prompt.replace(",", " ").replace("?", " ").split():
        try:
            move = board.parse_san(word)
            safe_ucis.append(move.uci())
        except ValueError:
            try:
                move = chess.Move.from_uci(word)
                if move in board.legal_moves:
                    safe_ucis.append(move.uci())
            except ValueError:
                pass
                
    # 2. Maia-3 Preprocessing (Using our recent fixes!)
    board_tokens = tokenize_board(board)
    maia_cfg = MaiaConfig(history=coach.config.maia.history)
    
    historical_tokens_tensor = get_historical_tokens(
        board_history=[board_tokens], 
        cfg=maia_cfg, 
        base=0.0, 
        inc=0.0, 
        clk_left_before=0.0, 
        clk_ponder=0.0
    )
    
    # 3. Add Batch Dimensions and move to GPU
    # Cast to float because get_historical_tokens returns a PyTorch tensor directly
    maia_input_tokens = historical_tokens_tensor.float().unsqueeze(0).to(device)
    self_elos_tensor = torch.tensor([self_elo], dtype=torch.float32, device=device)
    oppo_elos_tensor = torch.tensor([oppo_elo], dtype=torch.float32, device=device)
    
    # 4. Generate the Response!
    responses = coach.generate(
        boards=[board],
        maia_tokens=maia_input_tokens,
        self_elos=self_elos_tensor,
        oppo_elos=oppo_elos_tensor,
        user_messages=[user_prompt],
        queried_ucis=[safe_ucis] if safe_ucis else [None],
        max_new_tokens=max_new_tokens,
        temperature=temperature,
    )
    
    return responses[0]
```
